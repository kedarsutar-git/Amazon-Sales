##### Gathering Data as csv file

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
df['rating'].value_counts()
```

```python
df.query('rating == "|"')
```

```python
# Changing Rating Columns Data Type

df['rating'] = df['rating'].str.replace('|', '3.9').astype('float64')
```

```python
# Changing 'rating_count' Column Data Type

df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
df.info()
```

```python
df[df["rating_count"].isnull()]
```

```python
df.fillna("Data not available",inplace=True)
```

### Data Visualization

```python
plt.scatter(df["actual_price"],df["rating"],color = "red",marker = "+")
plt.title("Actual Price vs Rating")
plt.xlabel("Actual_price")
plt.ylabel("Rating")
```

```python
plt.scatter(df["discounted_price"],df["rating"],color="red",marker="+")
plt.title("Discounted Price vs rating")
plt.xlabel("Discounted price")
plt.ylabel("Rating")
```

```python
means = [3125.31, 5444.99, 0.4769, 4.096]
columns = ['Discounted Price', 'Actual Price', 'Discount %', 'Rating']
plt.bar(columns, means)
plt.title('Mean Value Comparison')
plt.ylabel('Mean')
plt.xticks(rotation='vertical')
plt.show()
```

```python
minimum = [39,39,0,0]
column = ["discounted price","actual price","discounted percentage","rating"]
plt.bar(column,minimum)
plt.title("minmum value comparison")
plt.ylabel("minimum value")
plt.xticks(rotation='vertical')
plt.show()
```

```python
std = [6944.304394,10874.826864,0.216359,0.291620]
column = ["discounted price","actual price","discounted perca=entage","rating"]
plt.xticks(rotation='vertical')
plt.title("std value comparison")
plt.ylabel("std")
plt.bar(column,std)
plt.show()
```

```python
Max = [77990.000000,139900.000000,0.940000,5.000000]
column = ["discounted price","actual price","discounted percentage","rating"]
plt.bar(column,Max)
plt.title("maximum value comparison")
plt.xticks(rotation = "vertical")
plt.ylabel("Maximum")
plt.show()
```
